## Spritmonitor data preprocessing

```python
def brand_splitter(row):
    return row['Car Name'].split('-')[0]
def model_splitter(row):
    return row['Car Name'].split('-')[1]
```

```python
spritmonitor_data['Brand'] = spritmonitor_data.apply(brand_splitter, axis=1)
```

```python
spritmonitor_data['Model'] = spritmonitor_data.apply(model_splitter, axis=1)
```

```python
spritmonitor_data.head()
```

```python
spritmonitor_data = spritmonitor_data.drop(columns=['Car Name'])
```

```python
spritmonitor_data.head()
```

```python
spritmonitor_data = spritmonitor_data[['Brand', 'Model', 'CO2']]
spritmonitor_data.head()
```

## Honestjohn data preprocessing

```python
honestjohn_data.head()
```

```python
def mpg_calc(row):
    if row['MPG'] == '–' or row['MPG'] == '-':
        return float('NaN')
    
    elif row['MPG'].__contains__('–'):
        mpg_low = int(row['MPG'].split('–')[0])
        mpg_high = int(row['MPG'].split('–')[1])
        return (mpg_low + mpg_high) / 2
    
    else:
        return int(row['MPG'])
```

```python
honestjohn_data['avg_mpg'] = honestjohn_data.apply(mpg_calc, axis=1)
honestjohn_data.head()
```

```python
honestjohn_data.dropna(inplace=True)
honestjohn_data.reset_index(inplace=True)
honestjohn_data.drop(columns=['MPG'])
```

```python
honestjohn_data.drop(columns=['index'], inplace=True)
```

```python
honestjohn_data.drop(columns=['MPG'], inplace=True)
```

```python
honestjohn_data
```

```python
def model_manipulator(row):
    model = row['Model'].split('-')[0]
    return model
```

```python
honestjohn_data['Model'] = honestjohn_data.apply(model_manipulator, axis = 1)
```

```python
honestjohn_data
```

```python
def brand_manipulator(row):
    left = row['Brand'].split(' (')[0]
    brand = left.split()[0]
    if brand == 'Aston':
        return 'Aston Martin'
    elif brand == 'Alfa':
        return 'Alfa Romeo'
    elif brand == 'Land':
        return 'Land Rover'
    else:
        return brand 
```

```python
honestjohn_data['Brand'] = honestjohn_data.apply(brand_manipulator, axis=1)
```

```python
honestjohn_data
```

```python
honestjohn_data['CO2'] = 7100 / honestjohn_data['avg_mpg']
```

```python
honestjohn_data.drop(columns=['avg_mpg'], inplace=True)
```

```python
honestjohn_data
```
